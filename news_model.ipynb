```python
import pandas as pd
data=pd.read_csv('final_news.csv')
data
```

```python
date=data.pop('Date')
date
```

```python
daten=[]
from functools import reduce
for d in date:
    st=str(d)
    li=st.split("-")
    li.reverse()
    lis = reduce(lambda a, b : a+ "-" +str(b),li)
    daten.append(lis)
daten
```

```python
data
```

```python
data.insert(column='Date',value=daten,loc=0)
```

```python
data
```

```python
data.to_csv('news_final.csv')
```

```python
news= pd.DataFrame(columns=['Date','output'])
news
```

```python
i=0
while i<data.shape[0]:
    date=data['Date'][i]
    list=[]
    for j in range(i,data.shape[0]):
        if date == data['Date'][j]:
            list.append(data['output'][j])
        else:
            i=j-1
            break
    total=len(list)
    p0=list.count(0)/total
    p1=list.count(1)/total
    p2=list.count(2)/total
    max=0
    if p0>=p1 and p0>=p2:
        max=0
    elif p2>=p0 and p2>=p1:
        max=2
    elif p1>=p0 and p1>=p2:
        if p0==0 and p2==0:
            max=1
        elif p1!=0 and p1>p2:
            max=1
        elif p1!=0 and p1<p2:
            max=2
        elif p2!=0 and p2>p1:
            max=2
        elif p2!=0 and p1>p2:
            max=1
    news.loc[len(news.index)] = [date,max]
    i=i+1
```
